```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def func(t, R, L):
    global U0
    global C
    a = R/(2*L)
    b = (a**2 - 1/(L*C))**.5
    return((U0/(b*L))*np.exp(-a*t)*(np.exp(b*t)-np.exp(-b*t))/2)
```

```python
ri = 99.7 + 85 + 50 + 230
li = 36e-3
C = 1.652e-9
U0 = 7.56
```

# Nájst také R, L, aby bol fit čo najlepší.

```python
ti = np.linspace(0, 1e-3, 10000)
```

```python
plt.scatter(ti, func(ti, ri, li), s=1)
```
