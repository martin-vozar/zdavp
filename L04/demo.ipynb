```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
d1 = pd.read_csv('Data2.csv')
```

```python
d1.head()
```

```python
d1['Datetime'] = d1['Datum']+' '+d1['Cas']
d1['Datetime'] = pd.to_datetime(d1['Datetime'], format='%d.%m.%Y %H:%M:%S')
```

```python
plt.figure(figsize=(9,5))
plt.grid(which="both", linestyle='-.')

plt.scatter(d1['Datetime'], d1['Data'], s=1)
plt.plot(d1['Datetime'], d1['Data'], linewidth=0.1)

plt.xlim(pd.Timestamp('2022-07-01 00:00:00'),
         pd.Timestamp('2022-07-31 23:45:00'))

plt.show()
```
