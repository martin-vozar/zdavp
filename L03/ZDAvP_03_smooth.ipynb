```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def norm(xi, mu, sigma):
        return(np.exp(-(xi-mu)**2/(2*sigma**2))/(sigma*(2*np.pi)**.5))
```

```python
n = 121
x = np.linspace(-n/2, n/2, n)
uni = np.array([1/n]*n)
nor = norm(x, 0, n/4)
nor = nor/np.sum(nor)
```

```python
plt.bar(x, uni, alpha=0.7)
plt.bar(x, nor, alpha=0.7)
plt.show()
```

```python
x = np.linspace(-np.pi, np.pi, 1000)
y = np.sin(x)+np.random.normal(0,1,len(x))
```

```python
plt.grid(which='both', linestyle='-.')
plt.scatter(x, np.sin(x), s=1)
plt.scatter(x,y, s=1)
plt.show()
```

```python
uy = np.convolve(y, uni)
uy = uy[int(np.floor(n/2)):-int(np.floor(n/2))]

ny = np.convolve(y, nor)
ny = ny[int(np.floor(n/2)):-int(np.floor(n/2))]
```

```python
uy = np.convolve(uy, uni)
uy = uy[int(np.floor(n/2)):-int(np.floor(n/2))]

plt.grid(which="both", linestyle='-.')
plt.scatter(x, np.sin(x), s=1)
plt.scatter(x,uy, s=1)
plt.show()
```

```python
ny = np.convolve(ny, nor)
ny = ny[int(np.floor(n/2)):-int(np.floor(n/2))]

plt.grid(which="both", linestyle='-.')
plt.scatter(x, np.sin(x),s=1)
plt.scatter(x,ny, s=1)
plt.show()
```
