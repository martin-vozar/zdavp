```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 1000)
y = np.sin(x)
```

```python
dx = x[1]-x[0]
```

```python
dy = np.diff(y)
one_diff = dy/dx
```

```python
ddy = np.diff(one_diff)
two_diff = ddy/dx
```

```python
plt.scatter(x, y, s=1, label="y=sin(x)")
plt.scatter(x[:-1], one_diff, s=1, label="y=d(sin(x))/dx")
plt.scatter(x[1:-1], two_diff, s=1, label="y=d^2(sin(x))/dx^2")

plt.legend()
plt.show()
```
