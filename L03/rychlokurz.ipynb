```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
```

```python
R = 100
L = [36e-3, 72e-3, 108e-3]
U0 = 10
def I(t, Ri, L, C):
    alpha = Ri/(2*L)
    beta = (alpha**2 - 1/(L*C))**.5
    return(
        U0*np.exp(-alpha*t)*(np.exp(beta*t)-np.exp(-beta*t)/(2*beta*L))
        # U * e^(-at)
    )
```

```python
tl = np.linspace(0, 2e-3, 10000)
for l in L:
    plt.scatter(tl, I(tl, R, l, 1.652e-9), s=1)
#plt.scatter(t, I, s=1)
```
