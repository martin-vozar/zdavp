```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
```

```python
df.describe()
```

```python
dfSet = df[ df['variety'] == "Setosa" ]
dfVir = df[ df['variety'] == "Virginica"] 
dfVer = df[ df['variety'] == "Versicolor"]
```

b, r, g, m, k

```python
plt.scatter(dfSet['sepal.length'], dfSet['sepal.width'], c='m')
plt.scatter(dfVer['sepal.length'], dfVer['sepal.width'], c='k')
plt.scatter(dfVir['sepal.length'], dfVir['sepal.width'], c='g')

plt.show()
```

```python
from scipy.optimize import curve_fit
```

```python
def lin(x, a, b):
    return(a*x + b)
```

```python
xSet = np.linspace(np.min(dfSet['sepal.length']), np.max(dfSet['sepal.length']), 100)
xVir = np.linspace(np.min(dfVir['sepal.length']), np.max(dfVir['sepal.length']), 100)
xVer = np.linspace(np.min(dfVer['sepal.length']), np.max(dfVer['sepal.length']), 100)

pSet, cSet = curve_fit(lin, dfSet['sepal.length'], dfSet['sepal.width'])
pVir, cVir = curve_fit(lin, dfVir['sepal.length'], dfVir['sepal.width'])
pVer, cVer = curve_fit(lin, dfVer['sepal.length'], dfVer['sepal.width'])

plt.scatter(dfSet['sepal.length'], dfSet['sepal.width'], c='m')
plt.scatter(dfVir['sepal.length'], dfVir['sepal.width'], c='g')
plt.scatter(dfVer['sepal.length'], dfVer['sepal.width'], c='k')

plt.scatter(xSet, lin(xSet, pSet[0], pSet[1]), s=1, c='m')
plt.scatter(xVir, lin(xVir, pVir[0], pVir[1]), s=1, c='g')
plt.scatter(xVer, lin(xVer, pVer[0], pVer[1]), s=1, c='k')
```

```python
def kvad(x, a, b, c):
    return(a*x**2 + b*x + c)
```

```python
pSet, cSet = curve_fit(kvad, dfSet['sepal.length'], dfSet['sepal.width'])
pVir, cVir = curve_fit(kvad, dfVir['sepal.length'], dfVir['sepal.width'])
pVer, cVer = curve_fit(kvad, dfVer['sepal.length'], dfVer['sepal.width'])

plt.scatter(dfSet['sepal.length'], dfSet['sepal.width'], c='m')
plt.scatter(dfVir['sepal.length'], dfVir['sepal.width'], c='g')
plt.scatter(dfVer['sepal.length'], dfVer['sepal.width'], c='k')

plt.scatter(xSet, kvad(xSet, pSet[0], pSet[1], pSet[2]), s=1, c='m')
plt.scatter(xVir, kvad(xVir, pVir[0], pVir[1], pVir[2]), s=1, c='g')
plt.scatter(xVer, kvad(xVer, pVer[0], pVer[1], pVer[2]), s=1, c='k')
```
